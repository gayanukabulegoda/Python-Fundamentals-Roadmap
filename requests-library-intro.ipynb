```python
import requests

reponse = requests.get('https://api.github.com/events')
print(reponse)
print(reponse.status_code)
print(reponse.text)
```

```python
# .status_code - HTTP status code of the response
print(reponse.status_code)

if reponse.status_code == 200:
    print('Success')
elif reponse.status_code == 404:
    print('Not Found')
else:
    print('Error while fetching data')
```

```python
# .headers - HTTP headers of the response
print(reponse.headers)

content_type = reponse.headers['content-type']
print(content_type)
```

```python
# .text - Content of the response, in unicode
print(reponse.text)
```

```python
# .json() - Content of the response, in JSON
print(reponse.json())

if 'application/json' in reponse.headers['content-type']:
    data = reponse.json()
    print(data)
```
